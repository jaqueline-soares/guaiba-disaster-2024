```python
# In case of more than one observation for the same period, keep the last one
df = df[~df.datahora.duplicated('last')]

# Ensuring that the time series has an observation for every time period
df = pd.date_range(start = df.datahora.min(), end = df.datahora.max(), freq = '15Min').to_frame(index=False, name='datahora').merge(df, how='left')

# Separating level and precipitation series
df_nivel = df.loc[:, ~df.columns.str.startswith('chuva')]
df_chuva = df.loc[:, ~df.columns.str.startswith('nivel')]

# Replacing NAs in level dataframe by linear interpolation
df_nivel = df_nivel.interpolate(method='linear')

# Replacing NAs in precipitation dataframe by 0
df_chuva = df_chuva.fillna(0)

# Getting the 12h accumulated precipitation
df_chuva12h = df_chuva.rolling('12h', on='datahora', min_periods = 48).sum()
df_chuva12h.columns = df_chuva12h.columns.str.replace("chuva", "acum12h", regex=True)

# Joining level and 12h accumulated precipitation data
input = pd.merge(df_nivel, df_chuva12h, on='datahora',  how='left')

# Keeping only observations at full hours
input = pd.date_range(start = input.datahora.min().replace(minute=0), end = input.datahora.max(), freq = '1h').to_frame(index=False, name='datahora').merge(input, how='left')
input.head()
```

```python
# Creating the output variables 1, 6, 12, and 24 hours ahead
input['out1h'] = input['nivel_87450004'].shift(-1)
input['out6h'] = input['nivel_87450004'].shift(-6)
input['out12h'] = input['nivel_87450004'].shift(-12)
input['out24h'] = input['nivel_87450004'].shift(-24)
```

```python
# Splitting input and output data
output = input.loc[:, ~input.columns.str.startswith('nivel')]
output = output.loc[:, ~output.columns.str.startswith('acum12h')]
input = input.loc[:, ~input.columns.str.startswith('out')]
```

```python
# Keeping only observations from 03/nov/23 19:00:00 on
input = input.loc[(input['datahora'] >= '2023-11-03 18:45:00')].reset_index(drop = True)
input.head()
```

```python
# Keeping only observations until 05/jul/24 13:00:00
input = input.loc[(input['datahora'] <= '2024-07-05 13:00:00')].reset_index(drop = True)
input.tail()
```

```python
# Saving a csv with the full input dataset
input_full = input.round(2)
input_full.to_csv("output\\input_full_interpolated.csv", index = False)
```

```python
# Generating the input and output datasets for 1-step ahead forecasts
output = output.loc[(output['datahora'] >= '2023-11-03 18:45:00')].reset_index()
output1h = output[['index', 'datahora', 'out1h']]
output1h = output1h.dropna()

input1h = input.loc[(input['datahora'] < '2024-05-07 13:00:00')]

# Rounding to 2 decimal places
output1h = output1h.round(2)
input1h = input1h.round(2)

# Saving to csv
input1h.to_csv("output\\input1h_interpolated.csv", index = False)
output1h.to_csv("output\\output1h_interpolated.csv", index = False)
```
